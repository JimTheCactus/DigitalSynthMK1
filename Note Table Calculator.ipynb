###Note Table Generator###

```python
"""
Calculates the necessary period between states
for producing an appropriate note on a chromatic
scale within a given range of notes

"""

# Reference Constants
ref_note = 57.0
ref_freq = 440.0
semitones_per_octave=12.0
min_code = 0.0
max_code = 1023.0
cpu_freq = 8000000.0
timer_prescalar = 8.0

# Config Values
min_note = 24.0+12.0
max_note = min_note + semitones_per_octave * 2.0

# Calculated constants
codes = max_code-min_code
notes = max_note-min_note
range_scaler = notes/codes
timer_period = timer_prescalar/cpu_freq

result = """
#include <avr/pgmspace.h>

const PROGMEM unsigned int notes[] = {
"""

periods = []
for i in range(int(min_code),int(max_code)+1):
    note = round(i * range_scaler + min_note);
    freq = ref_freq * pow(2,(note-ref_note)/12);
    period = 1/freq;
    pulse_period = period/2;
    pulse_duration = int(round(pulse_period/timer_period))
    result += "    " + str(pulse_duration)
    if (i != max_code):
        result += ","
    else:
        result += " "
    result += " //" + str(i) + ": " + str(note) + " @ " + str(freq) + " Hz\r\n"

result += """};

#define get_period(x) pgm_read_word(notes+x)

"""

print result
```
